```python
n = 5

# Upper part
for i in range(1, n + 1):
    print("  " * (n - i), end="")   # leading spaces
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()

# Lower part
for i in range(n - 1, 0, -1):
    print("  " * (n - i), end="")   # leading spaces
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```
