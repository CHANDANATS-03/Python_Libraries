```python
#WAP to find sring of s2 present in s1 or not
s1="aaaabbbccccddabcdecefghhhh"
s2="abcd"
def sub_string(s1,s2):
    for i in range(len(s1)):
        if s1[i:len(s2)+i]==s2:
            return True
    return False
s1="aaaabbbccccddabcdecefghhhh"
s2="abcd"
if sub_string(s1,s2):
    print("s2 is substring of s1")
else:
    print("s2 is not a substring of s1")
```
