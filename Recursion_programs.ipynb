```python
#WAP to find factoril of a number
def fact_num(n):
    if n==0:
        return 1
    else:
        return n*fact_num(n-1)
fact_num(int(input("num:")))
```

```python
#WAP to convert decial to binary
def dtb(n):
    if n>1:
        dtb(n//2)
    print(n%2,end="")
dtb(int(input("decinal number")))
```

```python
#WAP to convert binary to decial
def btd(n):
    if n == '':
        return 0
    return int(n[0]) * (2 ** (len(n) - 1)) + btd(n[1:])
print(btd(input("Enter binary number: ")))
```

```python
#WAP convert 1st char to capitalize
def capitalize_first(l1):
    res=[]
    if len(l1)==0:
        return res
    res.append(l1[0].title())
    return res + capitalize_first(l1[1:])
l1=["recursion","in","python","Is","awesome"]
print(capitalize_first(l1))
```
