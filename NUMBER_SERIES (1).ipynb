```python
n = input("enter: ")
num = int(input("Number: "))
res = ""

for i in range(num):
    ch = n[i % len(n)]   # repeat string if num > length
    if i % 2 == 0:       # even index → lowercase
        if 'A' <= ch <= 'Z':
            ch = chr(ord(ch) + 32)   # convert to lowercase
    else:                # odd index → uppercase
        if 'a' <= ch <= 'z':
            ch = chr(ord(ch) - 32)   # convert to uppercase

    if i == 0:
        res = ch
    else:
        res =res+" "+ch   # add space between chars

print(res)
```
