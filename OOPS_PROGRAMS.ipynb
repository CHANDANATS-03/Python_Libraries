```python
#multilevel inheritance example 3
class Grand_Parent:
    def __init__(self, gp_name, gp_age):
        self.gp_name = gp_name
        self.gp_age = gp_age
    def details(self):
        print(f"Grand Parent Name: {self.gp_name}\nGrand Parent Age: {self.gp_age}")
class Parent(Grand_Parent):
    def __init__(self, gp_name, gp_age, p_name, p_age):
        super().__init__(gp_name, gp_age)
        self.p_name = p_name
        self.p_age = p_age
    def details2(self):
        self.details()
        print(f"Parent Name: {self.p_name}\nParent Age: {self.p_age}")

class Child(Parent):
    def __init__(self, gp_name, gp_age, p_name, p_age, c_name, c_age):
        super().__init__(gp_name, gp_age, p_name, p_age)
        self.c_name = c_name
        self.c_age = c_age
    def details3(self):
        self.details2()
        print(f"Child Name: {self.c_name}\nChild Age: {self.c_age}")

c = Child("girijamma", 80, "sudha", 50, "chandu", 20)
c.details3()
```

```python
class SchoolMember:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def details(self):
        print(f"Name: {self.name}\nAge: {self.age}")


class Teacher(SchoolMember):
    def __init__(self, name, age, subject, salary):
        super().__init__(name, age)
        self.subject = subject
        self.salary = salary
    def details1(self):
        self.details()
        print(f"Teacher Subject: {self.subject}\nSalary: ₹{self.salary}")


class Student(SchoolMember):
    def __init__(self, name, age, grade, roll_no):
        super().__init__(name, age)
        self.grade = grade
        self.roll_no = roll_no
    def details1(self):
        self.details()
        print(f"Student Grade: {self.grade}\nRoll No: {self.roll_no}")


class Staff(SchoolMember):
    def __init__(self, name, age, role, shift):
        super().__init__(name, age)
        self.role = role
        self.shift = shift
    def details1(self):
        self.details()
        print(f"Staff Role: {self.role}\nShift: {self.shift}")


# Objects
t = Teacher("chandana", 40, "Mathematics", 50000)
s = Student("Ananya", 16, "10th",3)
st = Staff("Appu", 35, "Clerk", "Morning")

# Output
print("Teacher details")
t.details1()
print("\n")

print("Student Details")
s.details1()
print('\n')

print("\nStaff details")
st.details1()
```

```python
#Hybrid inheritance example 1 it is a combination of single,multi level, multiple, hierarchical inheritance
class A:
    def __init__(self,a):
        self.a=a
    def meth_a(self):
        print("details")
        print("a=",self.a)
class B(A):
    def __init__(self,a,b):
        super().__init__(a)
        self.b=b
    def meth_b(self):
        self.meth_a(a)
        print("b=",self.b)

class C(B):
    def __init__(self,a,b,c):
        super().__init__(a,b)
        self.c=c
    def meth_c(self):
        self.meth_b(B)
        print("c=",self.c)

class D(C,B,A):
    def __init__(self,a,b,c,d):
        super().__init__(a,b,c)
        B.__init__(a,b)
        A.__init__(a)
        self.d=d
    def meth_c(self):
        self.meth_c(C)
        print("d=",self.d)

class E(A):
    def __init__(self,a,e):
        super().__init__(a)
        self.e=e
    def meth_e(self):
        self.meth_a(a)
        print("e=",self.e)
    
    
    
```

```python
#Hybrid inheritance example 2
class vehicle:
    def __init__(self,brand,color):
        self.brand=brand
        self.color=color
    def info(self):
        print("brand name= ",self.brand)
        print("color  =",self.color) 
    

class car(vehicle):
    def __init__(self,brand,color,model,seats):
        super().__init__(brand,color)
        self.model=model
        self.seats=seats
    def info1(self):
        self.info()
        print("model name =",self.model)
        print("seats =",self.seats)
print("\033[1mexample for single-level inheritance.\033[0m")
print("\033[1mcar details.\033[0m")
obj_car=car("tata","white","nexon",5)
obj_car.info1()
    
class electric_car(car):
    def __init__(self,brand,color,model,seats,battery_volts):
        super().__init__(brand,color,model,seats)
        self.battery_volts=battery_volts
    def info2(self):
        self.info1()
        print("battery_volts =",self.battery_volts)
print("\033[1mexample for multi-leve inheritance.\033[0m")
print("\033[1melectric car details.\033[0m")
obj_ec=electric_car("tata","white","curve E V",5,"720v")
obj_ec.info2()

    
class hibrid_car(electric_car,car,vehicle):
    def __init__(self,brand,color,model,seats,battery_volts,fual,milage):
        super().__init__(brand,color,model,seats,battery_volts)
        car.__init__(self,brand,color,model,seats)
        vehicle.__init__(self,brand,color)
        self.fual=fual
        self.milage=milage
    def info_hc(self):
        self.info2()
        print("fual = ",self.fual)
        print("milage = ",self.milage)
print("\033[1mexample for multipul inheritance.\033[0m")
print("\033[1mhybrid car details.\033[0m")
obj_hc=hibrid_car("toyoto","white","innova hybrid",8,"720v","petrol",30)
obj_hc.info_hc()
    
class bike(vehicle):
    def __init__(self,brand,color,model,fual):
        super().__init__(brand,color)
        self.model=model
        self.fual=fual
    def info_bike(self):
        print("\033[1mbike details.\033[0m")
        self.info()
        print("model name =",self.model)
        print("fual =",self.fual)
print("\033[1mexample for Hierarchical inheritance.\033[0m")
obj_bike=bike("TVS","white","XL100","petrol")
obj_bike.info_bike()
    
    
```

```python
class Account:
    def __init__(self, acc_no, holder_name):
        self.acc_no = acc_no
        self.holder_name = holder_name

    def account_info(self):
        print(f"Account Number: {self.acc_no}")
        print(f"Account Holder: {self.holder_name}")
class Savings(Account):
    def __init__(self, acc_no, holder_name, balance):
        Account.__init__(self, acc_no, holder_name)
        self.balance = balance
    def savings_info(self):
        print(f"Savings Balance: ₹{self.balance}")
class Loan(Account):
    def __init__(self, acc_no, holder_name, loan_amount):
        Account.__init__(self, acc_no, holder_name)
        self.loan_amount = loan_amount
    def loan_info(self):
        print(f"Loan Amount: ₹{self.loan_amount}")
class Customer(Savings, Loan):
    def __init__(self, acc_no, holder_name, balance, loan_amount):
        Savings.__init__(self, acc_no, holder_name, balance)
        Loan.__init__(self, acc_no, holder_name, loan_amount)
    def customer_info(self):
        print("Customer Full Details ")
        self.account_info()
        self.savings_info()
        self.loan_info()
cust1 = Customer(101, "Rahul Sharma", 50000, 200000)
cust1.customer_info()
```

```python
#Create a class Battery and it should contains BatteryType, Capacity, and methods like Charging() and Discharging(). 
#Create a class Mobile, it should contains Name, Price, OS, Color, and methods like Netflix(), Calling(), Message(), PUBG(). 
#When each method is called, some amount of battery need to reduce, and it should update in current_capacity,
#and when we called plug_into_charge(), the battery capacity should increase. give python code using composition

class Battery:
    def __init__(self, battery_type, capacity):
        self.battery_type = battery_type
        self.capacity = capacity        # Max capacity
        self.current_capacity = capacity  # Current available charge

    def charging(self, amount):
        """Charge the battery by given amount"""
        if self.current_capacity < self.capacity:
            self.current_capacity += amount
            if self.current_capacity > self.capacity:
                self.current_capacity = self.capacity
        print(f"🔋 Charging... Current Capacity: {self.current_capacity}/{self.capacity}")

    def discharging(self, amount):
        """Discharge the battery by given amount"""
        if self.current_capacity >= amount:
            self.current_capacity -= amount
        else:
            print("⚠️ Battery too low! Please charge.")
        print(f"🔋 After usage, Current Capacity: {self.current_capacity}/{self.capacity}")


class Mobile:
    bat = Battery("Li-ion", 100)
    def __init__(self, name, price, os, color):
        self.name = name
        self.price = price
        self.os = os
        self.color = color
        self.battery = battery  # Composition (Mobile HAS a Battery)

    def netflix(self):
        print(f"{self.name} is playing Netflix 🎬")
        self.battery.discharging(15)

    def calling(self):
        print(f"{self.name} is making a Call 📞")
        self.battery.discharging(5)

    def message(self):
        print(f"{self.name} is sending a Message 💬")
        self.battery.discharging(2)

    def pubg(self):
        print(f"{self.name} is playing PUBG 🎮")
        self.battery.discharging(25)

    def plug_into_charge(self, amount):
        print(f"{self.name} is plugged into charge ⚡")
        self.battery.charging(amount)


# --------- Usage Example ----------
mobile = Mobile("Samsung Galaxy", 50000, "Android", "Black", bat)

mobile.netflix()
mobile.calling()
mobile.message()
mobile.pubg()

# Charging
mobile.plug_into_charge(30)
```

```python
def process_values(*args, **kwargs):
    total_sum = 0
    
    for val in args:
        if isinstance(val, int):
            total_sum += val
    
    int_only_dict = {}
    for key, value in kwargs.items():
        if isinstance(value, int):
            int_only_dict[key] = value

    print(" Sum of integer values:", total_sum)
    if int_only_dict:
        print(" Dictionary with integer values only:", int_only_dict)

process_values(10, "hello", 20.5, 30, a=5, b="world", c=15, d=3.14, e=40)
```
