```python
df['Age'].isna()

for col in df.columns:
    if df[col].isnull().any():
        print(col,"-->",df[col].isnull().sum()," null values")
    
```

```python
for col in ["IMDb","Runtime"]:
    df[col]=df[col].fillna(df[col].mean())
for col in ["Directors","Genres","Country","Language"]:
    df[col]=df[col].fillna('unknown')
```

```python
df
```

```python
df["Age"]=df['Age'].fillna(df["Age"].mode()[0])
```

```python
df.info()
```

```python
df['Rotten Tomatoes']=df['Rotten Tomatoes'].str.rstrip('%').astype('float') #we can not convert srting to int 
```

```python
df.info()
```

```python
df['Rotten Tomatoes']=df['Rotten Tomatoes'].fillna(df['Rotten Tomatoes'].mean()) / 10
df['Rotten Tomatoes']=df['Rotten Tomatoes'].round(1)
df
```

```python
df['IMDb']=df['IMDb'].round(1)
df['Runtime']=df['Runtime'].round(1)
```

```python
df.head()
```

```python
df['Type'].unique()
```

```python
df=df.drop(columns=['Type','Unnamed: 0','Rotten Tomatoes'])
```

```python
df.head()
df=df.drop(columns=['ID'])
```

```python
df.info()
```

```python
df.columns
```

```python
# melt()= unpivoting : means cols to rows or (wider to longer format)
pd.melt(df,['Title', 'Year', 'Age', 'IMDb','Directors', 'Genres', 'Country', 'Language', 'Runtime'],['Netflix', 'Hulu', 'Prime Video',
       'Disney+'])
pd.melt(df,id_vars=['Title'],value_vars=['Netflix', 'Hulu', 'Prime Video',
       'Disney+'])
```

```python
plt_df=pd.melt(df,id_vars=['Title','IMDb','Runtime'],value_vars=['Netflix', 'Hulu', 'Prime Video',
       'Disney+'],var_name="Platforms",value_name="Availability")
plt_df
```

```python
plt_df.columns
```

```python
plt_df=plt_df.loc[plt_df["Availability"]==1,['Title', 'IMDb', 'Runtime', 'Platforms']].reset_index(drop=True)
```
