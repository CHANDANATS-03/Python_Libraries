```python
import pandas as pd
```

```python
df=pd.read_csv("HR Data-2.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df['Department']=df['Department'].str.title()
```

```python
df.info()
```

```python
df=df.drop(columns=['Unnamed: 30','Unnamed: 29','Unnamed: 27','Unnamed: 25','Unnamed: 22','Unnamed: 21','EmployeeCount','MonthlyRate','DailyRate','HourlyRate','StandardHours'])
```

```python
df.info()
```

```python
age_mean=df['Age'].mean()
df['Age']=df['Age'].fillna(age_mean)
```

```python
# for col in ["Department","Education"]:
#     df[col]=df[col].str.strip()
df['Department']=df['Department'].str.strip()
```

```python
df.head()
```

```python
df['EmployeeNumber'].sort_values() #or
df=df.sort_values(by='EmployeeNumber',ascending=True).reset_index(drop=True)#by=[] we can give muliple column name
df
```

```python
# att_count=df['DistanceFromHome'].value_counts()
# att_count
# att_cnt=att_count.reset_index()
# att_cnt
# df=pd.merge(df,att_cnt,how='inner',on='DistanceFromHome')
# df
```

```python
df.rename(columns={'EnvironmentSatisfaction': 'EnvSatisfaction'}, inplace=True)
```

```python
df['Age']=df['Age'].round(0)
```

```python
l=['EducationField','Gender','MaritalStatus','JobRole']
for i in l:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
df['EmployeeNumber'].unique()
```

```python
df.info()
```

```python
df['EmployeeNumber'].duplicated().any()
```

```python
df['EmployeeNumber']=df['EmployeeNumber'].drop_duplicates(keep="last")
```

```python
df
```

```python
df.info()
```

```python
job=df['JobSatisfaction'].mean()
df['JobSatisfaction']=df['JobSatisfaction'].fillna(job)
```

```python
df=df.dropna(axis=0)
```

```python
df.info()
```

```python
df.agg({'YearsAtCompany':['min','max'],'Age':['min','max']})
```

```python
df['exp_level']=pd.cut(df['YearsAtCompany'],bins=[0,5,30,40],labels=['Fresher','Intermediate','Expert'],include_lowest=True)
```

```python
df['age_category']=pd.cut(df['Age'],bins=[17,25,50,60],labels=['Teen','Adult','Senior'])
```

```python
df['status']=pd.cut(df['Attrition'],bins=[0,0.1,1],labels=['Active','Terminated'],include_lowest=True)
```

```python
df.info()
```

```python
df
```

```python
df["total_experirnce"]=df["TotalWorkingYears"]+df["YearsAtCompany"]
```
